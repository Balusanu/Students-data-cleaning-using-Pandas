## Data cleaning - Students' data

Data cleaning steps that we are going to follow:
1) Deleting redundant columns
2) Renaming columns
3) Dropping duplicates
4) Remove NaN(null) values from the dataset
5) Check for some more transformations

```python
import pandas as pd
```

```python
pd.__version__
```

```python
data = pd.read_csv(r"C:\Users\balus\OneDrive\Desktop\Projects\Data cleaning - Pandas-student info\Students_info_data.csv")
```

```python
data.info() ## It is showing datatypes, no of non-null entries & various other info about each column
```

```python
'''1) Deleting redundant columns
'Unnamed: 0','TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild' - 
These are the redundant columns (not so important columns) that we are deleting'''
data.drop(columns=['Unnamed: 0','TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild'],inplace=True)
```

```python
data
```

```python
'''2) Renaming columns, we are going to rename column NrSiblings as Numberofsiblings '''
data.rename(columns={"NrSiblings":"Noofsiblings"},inplace=True)
```

```python
data
```

```python
'''3) Dropping duplicates - we are going to check how many duplicate records are there & delete them'''
data.duplicated()
```

```python
data.duplicated().sum() ## There are 9 duplicate rows in the data
```

```python
data.drop_duplicates(inplace=True) ## deleting those duplicate rows
```

```python
data.duplicated().sum() ## now there are no duplicate rows
```

```python
'''4) Remove NaN(null) values from the dataset - we are going to find Null values & remove them'''
data.isna() ## Showing True wherever there is null values & false wherever there non-null values
```

```python
data.isna().sum() ## Shows No of null values column wise
```

```python
data.isna().sum().sum() ##Shows total No of null values across all columns
```

```python
data.dropna(inplace=True) ## removing null records & saving it
```

```python
data.isna().sum().sum() ## now there are no null values
```

```python
## 5) Check for some more transformations
data["Gender"].unique() ## checking unique entries in Gender column
```

```python
## Changing the gender values to 1 for female & 0 for male
data["Gender"] = data["Gender"].apply(lambda x:1 if x == "female" else 0)
```

```python
data ## we can see Index is not right order
```

```python
data.reset_index(inplace=True)
```

```python
data
```

```python
data.drop(columns="index",inplace=True)
```

```python
data
```

```python
type(data["Noofsiblings"][0]) ## checking the datatype of the column Noofsiblings
```

```python
data["Noofsiblings"] =  data["Noofsiblings"].astype(int)
```
